```python
import pandas as pd
import json
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

# Load the dataset
df = pd.read_csv('feature_vectors_syscallsbinders_frequency_5_Cat.csv')

# Adjust class labels to start from 0
X = df.drop('Class', axis=1)
y = df['Class'] - 1  # Shift labels: [1,2,3,4,5] â†’ [0,1,2,3,4]

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)

#  clean version without warning
clf = XGBClassifier(eval_metric='mlogloss')

# Train XGBoost model
clf = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
clf.fit(X_train, y_train)

# When training the model
expected_features = list(X.columns)
with open("expected_features.json", "w") as f:
    json.dump(expected_features, f)

# Predict and evaluate
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
# Cross Validatoin
# Look into various evaluation metrics that can be used for XGBoost i.e., ROC curve, Cross validation.
# is the data imbalanced ? 
# use of unsupervised Learning !
#  Generate data using JSON file/share the file.
# Why is the result so good.

from sklearn.model_selection import cross_val_score

cv_result = np.mean(cross_val_score(clf, X, y))
cv_result
```

```python
# consumer_predictor.py
from kafka import KafkaConsumer
import json
import pandas as pd
import joblib
import numpy as np

# Load the trained model
model = joblib.load("xgboost_model.pkl")
#model = joblib.load("lightgbm_model.pkl")
# Create Kafka Consumer
consumer = KafkaConsumer(
    'android_features',
    bootstrap_servers='localhost:9092',
    auto_offset_reset='earliest',
    enable_auto_commit=True,
    group_id='malware-detector-group',
    value_deserializer=lambda x: json.loads(x.decode('utf-8'))
)

print("[Consumer] Waiting for messages...")

# Listen and predict
for message in consumer:
    feature_dict = message.value
    #print(f"[Consumer] Received: {feature_dict}")

    # Convert dict to DataFrame
    X = pd.DataFrame([feature_dict])
    #X = pd.DataFrame([feature_dict])
    X = X.drop(columns=["Class"], errors="ignore")  # this safely removes 'Class' if it exists



    
    # Ensure correct column order (should match training)
    # If your model was trained with a specific column order, make sure to apply that
    # X = X[trained_column_order]

    prediction = model.predict(X)[0]
    label = "Malignant" if prediction == 1 else "Benign"

    #print(f"[Prediction] The app is likely: {label}")
    print(f"Prediction: {label}")
```
